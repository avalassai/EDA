```python
df_country = pd.read_excel('country-code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head()
```

```python
final_df.Country.value_counts()
```

```python
Country_values = final_df.Country.value_counts().values
Country_names = final_df.Country.value_counts().index
```

```python
plt.pie(Country_values,labels=Country_names)
```

```python
plt.pie(Country_values[:3],labels=Country_names[:3],autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Score'})
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y='Rating Score',data=ratings,hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
```

```python
rate = final_df.groupby(['Rating text']).size().reset_index().rename(columns={0:"Rating text count"})
rate
```

```python
sns.barplot(x="Rating text",y="Rating text count",data=rate)
```

```python
sns.countplot(x="Rating color",data=ratings ,palette=['blue','red','orange','yellow','green','green'])
```
